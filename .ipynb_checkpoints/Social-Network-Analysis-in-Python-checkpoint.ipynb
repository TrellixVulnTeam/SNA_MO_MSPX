```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
import collections
import math
from math import *
f = pd.read_csv("1.tsv", sep='\t')
G1 = nx.from_pandas_edgelist(f, source='Partner 1', target='Partner 4',edge_attr=None, create_using=nx.Graph())
```

```python
print(nx.info(G1))#all information
print(G1.size())#network size
```

# Betweenness Centrality

```python
pos = nx.spring_layout(G1)
betCent = nx.betweenness_centrality(G1, normalized=True, endpoints=True)
node_color = [20000.0 * G1.degree(v) for v in G1]
node_size =  [v * 10000 for v in betCent.values()]
plt.figure(figsize=(20,20))
nx.draw_networkx(G1, pos=pos, with_labels=False,
                 node_color=node_color,
                 node_size=node_size )
plt.axis('on')
sorted(betCent, key=betCent.get, reverse=True)[:50]
#print(pos)
print(betCent)
```

# Degree Centrality

```python
pos = nx.spring_layout(G1)
degCent = nx.degree_centrality(G1)
node_color = [20000.0 * G1.degree(v) for v in G1]
node_size =  [v * 10000 for v in degCent.values()]
plt.figure(figsize=(15, 15))
nx.draw_networkx(G1, pos=pos, with_labels=False,
                 node_color=node_color,
                 node_size=node_size )
plt.axis('on')

sorted(degCent, key=degCent.get, reverse=False)[:1000]
print(degCent)
```

```python
sorted(degCent, key=degCent.get, reverse=True)[:10]
```

# Closeness Centrality

```python
pos = nx.spring_layout(G1)
cloCent = nx.closeness_centrality(G1)
node_color = [20000.0 * G1.degree(v) for v in G1]
node_size =  [v * 1000 for v in cloCent.values()]
plt.figure(figsize=(15,15))
nx.draw_networkx(G1, pos=pos, with_labels=False,
                 node_color=node_color,
                 node_size=node_size )
plt.axis('on')
sorted(cloCent, key=cloCent.get, reverse=False)[:1000]
print(cloCent)
```

```python
import collections
degree = G1.degree()
degree_sequence = sorted([d for n, d in G1.degree()], reverse=False)  # degree sequence
degreeCount = collections.Counter(degree_sequence)
deg, cnt = zip(*degreeCount.items())

fig, ax = plt.subplots()
plt.bar(deg, cnt, width=10, color="c")

plt.title("Degree Histogram")
plt.ylabel("Count")
plt.xlabel("Degree")
ax.set_xticks([d + 0.4 for d in deg])
ax.set_xticklabels(deg)

# draw graph in inset
plt.axes([0.4, 0.4, 0.5, 0.5])
Gcc = G1.subgraph(sorted(nx.connected_components(G1), key=len, reverse=True)[0])
pos = nx.spring_layout(G1)

plt.axis("off")
nx.draw_networkx_nodes(G1, pos, node_size=30)
nx.draw_networkx_edges(G1, pos, alpha=0.4)
plt.show()
```
